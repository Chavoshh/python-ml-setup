import torch
import numpy as np

data = [[1, 2],[3, 4]]

##shape is a tuple of tensor dimensions. In the functions below, it determines the dimensionality of the output tensor.
shape = (2,3)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} 'n")
print(f"Zeros Tensor: \n {zeros_tensor}")

##Tensor attributes describe their shape, data type, and the device on which are stored.
tensor = torch.rand(3, 4)
print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tnesor is stored on: {tensor.device}")

## We move our tensor to the GPU if available
device = torch.accelerator.current_accelerator().type if torch.accelerator.is_available() else 'cpu'
tensor = tensor.to(device)
print(f"Device tensor is stored on: {tensor.device}")

## Standard numpy-like indexing & slicing:

### Create a 3x4 tensor of random numbers:
x = torch.rand(3, 4)

### NumPy - Tensor conversion
arr = np.linspace(0, 10, 50)
tensor_from_np = torch.from_numpy(arr)
back_to_np = tensor_from_np.numpy()

print("\nOriginal NumPy array:")
print(arr[:5])
print("\nConverted tensor:")
print(tensor_from_np[:5])
print("\nBack to NumPy:")
print(back_to_np[:5])
print("\nAre they the same?")
print(np.array_equal(arr, back_to_np))

### Matrix multiplication
A = torch.rand(4, 3)
B = torch.rand(3, 5)

### (4 x 3) @ (3 x 5) -> (4 x 5)

print("\nMatrix mult. output shape prediction: (4, 5)")
C1 = torch.matmul(A, B)
C2 = A @ B
C3 = A.mm(B)

print("\nUsing torch.matmul:")
print(C1.shape)
print("\nUsing @ operator")
print(C2.shape)
print("\nUsing .mm():")
print(C3.shape)

### Element-wise vs. reduction operations
x = torch.tensor([1., 2., 3.])
print("\nTensor x:")

### Element-wise operations
print("\ntorch.sin(x):")
print(torch.sin(x))
print("\nx**2:")
print(x**2)

### Reduction operations
print("\nx.mean():")
print(x.mean())

print(x.std())

# A scalar function: y = x^2 + 3x + 5
