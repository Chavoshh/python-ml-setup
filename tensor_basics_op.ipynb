```python
import torch
import numpy as np
```

```python
### Create a 3x4 tensor of random numbers:
x = torch.rand(3, 4)
print("Original tensor (3x4):")
print(x)
print("Shape:", x.shape)
```

```python
### Reshape operations
x1 = x.reshape(6, 2)
print("\nReshaped to 6 x 2:")
print(x1)
print("Shape:", x1.shape)
x2 = x.reshape(2, 6)
print("\nReshaped to 2 x 6:")
print(x2)
print("Shape:", x2.shape)
```

```python
### Slicing
print("\nFirst row:")
print(x[0])

print("\nLast column:")
print(x[:, -1])

print("\nBottom-right 2x2 block:")
print(x[-2:, -2:])
```
