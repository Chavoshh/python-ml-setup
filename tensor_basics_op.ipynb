import torch
import numpy as np

### Create a 3x4 tensor of random numbers:
x = torch.rand(3, 4)
print("Original tensor (3x4):")
print(x)
print("Shape:", x.shape)

### Reshape operations
x1 = x.reshape(6, 2)
print("\nReshaped to 6 x 2:")
print(x1)
print("Shape:", x1.shape)
x2 = x.reshape(2, 6)
print("\nReshaped to 2 x 6:")
print(x2)
print("Shape:", x2.shape)

### Slicing
print("\nFirst row:")
print(x[0])

print("\nLast column:")
print(x[:, -1])

print("\nBottom-right 2x2 block:")
print(x[-2:, -2:])

### NumPy - Tensor conversion
arr = np.linspace(0, 10, 50)
tensor_from_np = torch.from_numpy(arr)
back_to_np = tensor_from_np.numpy()

print("\nOriginal NumPy array:")
print(arr[:5])
print("\nConverted tensor:")
print(tensor_from_np[:5])
print("\nBack to NumPy:")
print(back_to_np[:5])
print("\nAre they the same?")
print(np.array_equal(arr, back_to_np))

### Matrix multiplication
A = torch.rand(4, 3)
B = torch.rand(3, 5)

### (4 x 3) @ (3 x 5) -> (4 x 5)

print("\nMatrix mult. output shape prediction: (4, 5)")
C1 = torch.matmul(A, B)
C2 = A @ B
C3 = A.mm(B)

print("\nUsing torch.matmul:")
print(C1.shape)
print("\nUsing @ operator")
print(C2.shape)
print("\nUsing .mm():")
print(C3.shape)

### Element-wise vs. reduction operations
x = torch.tensor([1., 2., 3.])
print("\nTensor x:")
print(x)

### Element-wise operations
print("\ntorch.sin(x):")
print(torch.sin(x))
print("\nx**2:")
print(x**2)

### Reduction operations
print("\nx.mean():")
print(x.mean())
print("\nx.std():")
print(x.std())
